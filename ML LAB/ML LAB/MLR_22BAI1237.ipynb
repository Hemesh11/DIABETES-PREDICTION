```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
data=pd.read_csv('Student.csv')
print(data.head())
```

```python

from statsmodels.stats.outliers_influence import variance_inflation_factor 

data['Extra_Activities'] = data['Extra_Activities'].map({'Yes':1, 'No':0}) 
  
# the independent variables set 
X = data[['Hours_Studied', 'Previous_Scores', 'Extra_Activities','Sleep_Hours','Papers_Practiced']] 
  
# VIF dataframe 
vif_data = pd.DataFrame() 
vif_data["feature"] = X.columns 
  
# calculating VIF for each feature 
vif_data["VIF"] = [variance_inflation_factor(X.values, i) 
                          for i in range(len(X.columns))] 
  
print(vif_data)
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

formula = 'Performance_Index ~ Hours_Studied + Previous_Scores + Extra_Activities+ Sleep_Hours + Papers_Practiced'
model = ols(formula, data=data).fit()
anova_table = sm.stats.anova_lm(model)
print(anova_table)
```

```python
anova_results = sm.stats.anova_lm(model)

f_statistic = [anova_results['F'][i] for i in range(5)]
p_value = [anova_results['PR(>F)'][i] for i in range(5)]

print(f"F-statistic: {f_statistic}")
print(f"p-value: {p_value}")
```

```python
#hyothetical Testing 
hyp_test=False
count=0
for i in range(5):
    if p_value[i] < 0.05 :
        count+=1
if(count==5):
    hyp_test=True
if(hyp_test):
    print("We reject the null hypothesis H0")
    print("All 5 parameters are significant for Performance Index")
else:
    print("We fail to reject the null hypothesis H0")
```

```python
#R^2 and Adjusted R^2
formula = 'Performance_Index ~ Hours_Studied + Previous_Scores + Extra_Activities+ Sleep_Hours + Papers_Practiced'
model = ols(formula, data=data).fit()
model.summary()
```

```python

num_coeff = len(model.params)
for i in range(num_coeff):
    coef_values = [1 if j == i else 0 for j in range(num_coeff)]
    t_test_result = model.t_test(coef_values)
    
    print(f"T-Test for coefficient {i}:")
    print(t_test_result)
    print("\n")
```
