```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

# Generate linear dataset
np.random.seed(0)
X_linear = 2 * np.random.rand(100, 1)
y_linear = 4 + 3 * X_linear + np.random.randn(100, 1)

# Split data into training and testing sets
X_train_linear, X_test_linear, y_train_linear, y_test_linear = train_test_split(X_linear, y_linear, test_size=0.2, random_state=42)

# Train neural network on linear dataset
nn_linear = MLPRegressor(hidden_layer_sizes=(10,), activation='relu', solver='adam', max_iter=1000, random_state=42)
nn_linear.fit(X_train_linear, y_train_linear)

# Predict on test data
y_pred_linear = nn_linear.predict(X_test_linear)

# Plot linear dataset and predictions
plt.scatter(X_test_linear, y_test_linear, color='blue', label='Actual')
plt.scatter(X_test_linear, y_pred_linear, color='red', label='Predicted')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Dataset and Predictions')
plt.legend()
plt.show()

mse_linear = mean_squared_error(y_test_linear, y_pred_linear)
print("Mean Squared Error for linear dataset:", mse_linear)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

# Generate non-linear dataset (quadratic)
np.random.seed(0)
X_nonlinear = 2 * np.random.rand(100, 1)
y_nonlinear = 0.5 * X_nonlinear**2 + X_nonlinear + 2 + np.random.randn(100, 1)

# Split data into training and testing sets
X_train_nonlinear, X_test_nonlinear, y_train_nonlinear, y_test_nonlinear = train_test_split(X_nonlinear, y_nonlinear, test_size=0.2, random_state=42)

# Train neural network on non-linear dataset
nn_nonlinear = MLPRegressor(hidden_layer_sizes=(10,), activation='relu', solver='adam', max_iter=1000, random_state=42)
nn_nonlinear.fit(X_train_nonlinear, y_train_nonlinear)

# Predict on test data
y_pred_nonlinear = nn_nonlinear.predict(X_test_nonlinear)

# Plot non-linear dataset and predictions
plt.scatter(X_test_nonlinear, y_test_nonlinear, color='blue', label='Actual')
plt.scatter(X_test_nonlinear, y_pred_nonlinear, color='red', label='Predicted')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Non-linear Dataset and Predictions')
plt.legend()
plt.show()

# Calculate Mean Squared Error
mse_nonlinear = mean_squared_error(y_test_nonlinear, y_pred_nonlinear)
print("Mean Squared Error:", mse_nonlinear)
```
