```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans

# Load the Iris dataset
iris = load_iris()
iris_data = pd.DataFrame(data=np.c_[iris['data'], iris['target']],
                         columns=iris['feature_names'] + ['target'])

# Extract the features (excluding the target)
data = iris_data.iloc[:, :-1].values

# Manually implement KMeans++ initialization for the first n centroids
def kmeans_plus_plus(X, n_clusters):
    centroids = [X[np.random.randint(X.shape[0])]]  # Step 1

    while len(centroids) < n_clusters:
        # Step 2: Compute the distance from each point to the nearest centroid
        distances = np.array([min([np.linalg.norm(x - c) for c in centroids]) ** 2 for x in X])
        probabilities = distances / distances.sum()  # Step 3: Compute probability distribution
        cum_probabilities = probabilities.cumsum()
        r = np.random.rand()
        for j, p in enumerate(cum_probabilities):
            if r < p:
                centroids.append(X[j])  # Step 4: Choose a new centroid
                break

    return np.array(centroids)

# Perform KMeans++ initialization
initial_centroids = kmeans_plus_plus(data, 3)  # Change '3' to the desired number of clusters

# Perform KMeans clustering with the initialized centroids
kmeans = KMeans(n_clusters=len(initial_centroids), init=initial_centroids, n_init=1)
kmeans.fit(data)

# Plot the data points, centroids, and centroid coordinates
plt.figure(figsize=(8, 6))
plt.scatter(data[:, 0], data[:, 1], c=kmeans.labels_, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', s=200, c='red')
plt.title(f'KMeans++ Clustering with {len(initial_centroids)} clusters')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

silhouette_score and elbow method of evaluation 

```python

from sklearn.metrics import silhouette_score
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris

def evaluate_kmeans(X):
    silhouette_scores = []
    sse = []
    K = range(2, 11)

    for k in K:
        kmeans = KMeans(n_clusters=k)
        kmeans.fit(X)
        
        # Silhouette coefficient
        silhouette_score_val = silhouette_score(X, kmeans.labels_)
        silhouette_scores.append(silhouette_score_val)
        
        # Sum of squared errors (SSE)
        sse_val = kmeans.inertia_
        sse.append(sse_val)
        
        print(f"k={k}, Silhouette Score={silhouette_score_val}, SSE={sse_val}")


    # Plot the Silhouette scores as a histogram
    plt.figure(figsize=(8, 5))
    plt.bar(range(2, 11), silhouette_scores)
    plt.xlabel('Number of Clusters')
    plt.ylabel('Silhouette Score')
    plt.title('Silhouette Score for Different Cluster Numbers')
    plt.show()

    # Plot SSE
    plt.figure(figsize=(10, 5))
    plt.plot(K, sse, marker='o', color='orange')
    plt.xlabel('Number of Clusters (k)')
    plt.ylabel('Sum of Squared Errors (SSE)')
    plt.title('Elbow Method using Sum of Squared Errors (SSE)')
    plt.grid(True)
    plt.show()

# Load the Iris dataset
iris = load_iris()
X = iris.data  # Using all features of the Iris dataset

evaluate_kmeans(X)
```
